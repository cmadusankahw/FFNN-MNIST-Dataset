```python
from sklearn.datasets import load_iris
iris=load_iris()

data=iris.data
target=iris.target

from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
#print(test_target)
```

```python
from keras.layers import Dense
from keras.models import Sequential

model=Sequential()

model.add(Dense(8,input_dim=4,activation='relu'))
#1st hidden layer

model.add(Dense(8,input_dim=8,activation='relu'))
#2nd hidden layer

model.add(Dense(3,input_dim=8,activation='softmax'))#output layer
#to get result as a presentage/probability

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy']) #sequential gradient decent
#cross_entropy bcz of classification problem



```

```python
print(train_target)
```

```python
from keras.utils import np_utils

train_target=np_utils.to_categorical(train_target)

print(train_target)
```

```python
history=model.fit(train_data,train_target,epochs=500)
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['loss'])
plt.show()


plt.plot(history.history['acc'])
plt.show()
```

```python
print(test_target)

import numpy as np
result=model.predict(test_data)
#print(result)
print(np.max(result,axis=1))
```

```python
print(np.argmax(result,axis=1))
```
