```python
from keras import layers,models

model=models.Sequential()
#model is the NN

model.add(layers.Dense(8,input_dim=1,activation='linear')) #fully conneted layer.. no need to add input layer.it will gnenrate automatically
#this is the first hidden layer this has 12 nurones

model.add(layers.Dense(1,input_dim=8,activation='linear'))
#output layer

model.compile(loss='mse',optimizer='adam')

import numpy as np

x=np.arange(0,10,0.2)
y=(x*np.power(x,2))+(3*x)+4

model.fit(x,y,epochs=1000)

result=model.redict([0.5])
print(result)
```
